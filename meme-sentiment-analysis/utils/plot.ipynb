```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 900
```

```python
names = ['TF-IDF + SVM', 'ResNet50', 'BERT', 'roBERTa', 'ResNet152', 'InceptionV3', 'VGG16', 'VGG19', 'EfficientNetV2B0', 
        'EfficientNetV2M', 'EfficientNetV2B0 + BERT + RF', 'EfficientNetV2B0 + BERT + SVM', 'Multimodal', 'EfficientNetV2B0 + roBERTa + RF', 'EfficientNetV2B0 + roBERTa + SVM']

types = ['Text','Image','Text','Text','Image','Image','Image','Image','Image','Image','Ensemble','Ensemble', 'Multimodal', 'Ensemble', 'Ensemble']

bal_accs = [0.3215, 0.2995, 0.3130, 0.3197, 0.3160, 0.3333, 0.3306,
         0.3158, 0.3463, 0.3187, 0.3327, 0.3318, 0.33098, 0.330,  0.3318]


d = {'classifier':names,
     'type':types,
     'balanced accuracy':bal_accs
    }

df_results = pd.DataFrame(data=d)
df_results.sort_values(by='balanced accuracy', ascending=True, axis=0, inplace=True)
df_results
```

```python
ax = sns.barplot(data=df_results, x='classifier', y='balanced accuracy', hue='type')
ax.plot([-1, df_results.shape[0]], [0.33, 0.33], 'k--')
ax.annotate("*randomized algorithm (0.33)", [0.256, 0.256],
                xytext=(-6, 50), textcoords='offset points', ha="left", va="center", size = 5)

ax.set_xlabel('')
ax.set_ylabel('Balanced accuracy - validation')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlim([-.5,14.5])
ax.legend(loc='upper left', bbox_to_anchor=(1,0.25))
```
